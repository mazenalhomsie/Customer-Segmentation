# Seasonality Analysis
In this notebook, we analyze the trips based on seasons (Winter, Spring, Summer, Autumn).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/non_cancelled_trips.csv')

# Convert date columns to datetime
df['departure_time'] = pd.to_datetime(df['departure_time'])
df['check_in_time'] = pd.to_datetime(df['check_in_time'])
df['session_start'] = pd.to_datetime(df['session_start'])

# Determine Trip Start Date (Priority: Flight Departure -> Hotel Check-In -> Session Start)
df['trip_start_date'] = df['departure_time'].fillna(df['check_in_time']).fillna(df['session_start'])

print("Date columns processed.")
print(df[['trip_start_date', 'departure_time', 'check_in_time']].head())
```

### Define Seasons

```python
def get_season(date):
    if pd.isna(date):
        return 'Unknown'
    
    month = date.month
    # Northern Hemisphere Seasons
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['season'] = df['trip_start_date'].apply(get_season)

print("Season counts:")
print(df['season'].value_counts())
```

### Visualize Trips per Season

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='season', order=['Spring', 'Summer', 'Autumn', 'Winter'])
plt.title('Number of Trips per Season')
plt.xlabel('Season')
plt.ylabel('Number of Trips')
plt.show()
```

### Average Spend per Season (Flight Base Fare)

```python
if 'base_fare_usd' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df, x='season', y='base_fare_usd', order=['Spring', 'Summer', 'Autumn', 'Winter'], estimator='mean')
    plt.title('Average Flight Spend (Base Fare) per Season')
    plt.ylabel('Avg Base Fare (USD)')
    plt.show()
```

### Month-wise Distribution

```python
df['month'] = df['trip_start_date'].dt.month_name()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='month', order=month_order)
plt.title('Trips by Month')
plt.xticks(rotation=45)
plt.show()
```
