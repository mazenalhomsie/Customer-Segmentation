# Flight Spend Analysis (Per Seat)
In this section, we analyze the cost of flights per seat.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load non-cancelled trips data
df = pd.read_csv('../data/non_cancelled_trips.csv')
print(f"Loaded {len(df)} non-cancelled trips")

# Filter for flight bookings only
flight_bookings = df[df['flight_booked'] == True].copy()
print(f"Found {len(flight_bookings)} flight bookings")

# Check for missing or zero seats to avoid division by zero
flight_bookings = flight_bookings[flight_bookings['seats'] > 0]

# Calculate spend per seat
# base_fare_usd covers the trip (assuming it covers all seats? Usually base fare is total, but we should verify if it's per seat or total. 
# Typically in travel data, price might be total. Let's assume user wants Total / Seats.)
flight_bookings['spend_per_seat'] = flight_bookings['base_fare_usd'] / flight_bookings['seats']

# Display summary statistics
print("\n--- Flight Spend Per Seat (USD) ---")
print(flight_bookings['spend_per_seat'].describe())

print("\n--- Total Base Fare ---")
print(flight_bookings['base_fare_usd'].describe())
```

### Visualization of Flight Spend per Seat

```python
plt.figure(figsize=(12, 6))
sns.histplot(flight_bookings['spend_per_seat'], bins=50, kde=True)
plt.title('Distribution of Flight Spend Per Seat (USD)')
plt.xlabel('Price per Seat (USD)')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x=flight_bookings['spend_per_seat'])
plt.title('Boxplot of Flight Spend Per Seat')
plt.xlabel('Price per Seat (USD)')
plt.show()
```

### Average Spend per Seat by Airline

```python
avg_airline_price = flight_bookings.groupby('trip_airline')['spend_per_seat'].mean().sort_values(ascending=False)
print(avg_airline_price)

plt.figure(figsize=(12, 6))
avg_airline_price.plot(kind='barh')
plt.title('Average Spend Per Seat by Airline')
plt.xlabel('Price per Seat (USD)')
plt.show()
```
