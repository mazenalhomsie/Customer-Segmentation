# Hotel Spend Analysis (Per Night)
In this section, we analyze the cost of hotel stays per night for non-cancelled trips.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load non-cancelled trips data
df = pd.read_csv('../data/non_cancelled_trips.csv')
print(f"Loaded {len(df)} non-cancelled trips")

# Filter for hotel bookings only
hotel_bookings = df[df['hotel_booked'] == True].copy()
print(f"Found {len(hotel_bookings)} hotel bookings")

# Check for missing or zero nights to avoid division by zero
hotel_bookings = hotel_bookings[hotel_bookings['nights'] > 0]

# Calculate total hotel spend (base cost)
# Assuming hotel_per_room_usd is the base price per room per night
hotel_bookings['total_hotel_cost'] = hotel_bookings['hotel_per_room_usd'] * hotel_bookings['nights'] * hotel_bookings['rooms']

# Calculate Average Daily Rate (ADR) / Spend per Night
# If we have multiple rooms, is 'spend per night' per room or total?
# Usually ADR is per room. Total spend per night is total_cost / nights.
hotel_bookings['spend_per_night_total'] = hotel_bookings['total_hotel_cost'] / hotel_bookings['nights']
hotel_bookings['spend_per_night_per_room'] = hotel_bookings['hotel_per_room_usd']

# Display summary statistics
print("\n--- Hotel Spend Per Night (Per Room) ---")
print(hotel_bookings['hotel_per_room_usd'].describe())

print("\n--- Total Hotel Spend Per Night (All Rooms) ---")
print(hotel_bookings['spend_per_night_total'].describe())
```

### Visualization of Hotel Spend per Night

```python
plt.figure(figsize=(12, 6))
sns.histplot(hotel_bookings['hotel_per_room_usd'], bins=50, kde=True)
plt.title('Distribution of Hotel Price Per Room / Per Night (USD)')
plt.xlabel('Price (USD)')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x=hotel_bookings['hotel_per_room_usd'])
plt.title('Boxplot of Hotel Price Per Room / Night')
plt.xlabel('Price (USD)')
plt.show()
```

### Average Hotel Spend by Hotel Name (Top 10 Most Expensive)

```python
avg_hotel_price = hotel_bookings.groupby('hotel_name')['hotel_per_room_usd'].mean().sort_values(ascending=False).head(10)
print(avg_hotel_price)

plt.figure(figsize=(12, 6))
avg_hotel_price.plot(kind='barh')
plt.title('Top 10 Most Expensive Hotels (Avg Price Per Room/Night)')
plt.xlabel('Price (USD)')
plt.show()
```
