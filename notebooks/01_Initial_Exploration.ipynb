# Explorative Datenanalyse (EDA) - Woche 1

In diesem Notebook werden wir den TravelTide-Datensatz untersuchen. Ziel ist es, die Struktur der Daten zu verstehen, die Qualität zu prüfen und erste Erkenntnisse zu gewinnen.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Einstellungen für Plots
sns.set_theme(style="whitegrid")
pd.set_option('display.max_columns', None)
```

## 1. Daten laden
Wir laden die CSV-Dateien aus dem `data/` Ordner. Stellen Sie sicher, dass Sie `src/fetch_sample_data.py` ausgeführt haben.

```python
data_dir = '../data'

try:
    users = pd.read_csv(os.path.join(data_dir, 'users_sample.csv'))
    sessions = pd.read_csv(os.path.join(data_dir, 'sessions_sample.csv'))
    flights = pd.read_csv(os.path.join(data_dir, 'flights_sample.csv'))
    hotels = pd.read_csv(os.path.join(data_dir, 'hotels_sample.csv'))
    print("Daten erfolgreich geladen!")
except FileNotFoundError:
    print("Daten nicht gefunden. Bitte führen Sie 'src/fetch_sample_data.py' aus.")
```

## 2. Überblick über die Tabellen

```python
def print_overview(df, name):
    print(f"\n--- {name} ---")
    print(f"Shape: {df.shape}")
    print("\nInfo:")
    df.info()
    print("\nHead:")
    display(df.head())

print_overview(users, "Users")
print_overview(sessions, "Sessions")
print_overview(flights, "Flights")
print_overview(hotels, "Hotels")
```

## 3. Datenqualität prüfen
Prüfung auf fehlende Werte und Duplikate.

```python
for name, df in [('Users', users), ('Sessions', sessions), ('Flights', flights), ('Hotels', hotels)]:
    print(f"\n{name} - Fehlende Werte:")
    print(df.isnull().sum()[df.isnull().sum() > 0])
    print(f"{name} - Duplikate: {df.duplicated().sum()}")
```

## 4. Erste Analysen

### Nutzer-Demografie

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=users, x='gender')
plt.title('Geschlechterverteilung')
plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(data=users, x='home_country')
plt.title('Länderverteilung')
plt.show()
```

### Buchungsverhalten (Sessions)

```python
# Konvertiere Datum/Zeit
sessions['session_start'] = pd.to_datetime(sessions['session_start'])

# Buchungsrate
sessions['booked_any'] = sessions['flight_booked'] | sessions['hotel_booked']
booking_rate = sessions['booked_any'].mean() * 100
print(f"Allgemeine Buchungsrate: {booking_rate:.2f}%")

plt.figure(figsize=(10, 5))
sns.countplot(data=sessions, x='booked_any')
plt.title('Buchungen vs. Keine Buchungen')
plt.show()
```
